````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
````

````python
df_speed = pd.read_csv("../data/raw/time_dilation_high_speed_particles.csv")
df_gravity = pd.read_csv("../data/raw/gravitational_time_dilation.csv")
df_decay = pd.read_csv("../data/raw/particle_decay_time_dilation.csv")
````

## üîÅ Special Relativity ‚Äì Speed-Based Time Dilation

This plot shows how **time appears to dilate** (slow down) as an object's velocity approaches the **speed of light**.

We use the Lorentz factor (Œ≥) to calculate the observed time from Earth‚Äôs frame.

````python
plt.figure()
sns.lineplot(
    data=df_speed,
    x='Velocity Fraction of c',
    y='Dilated Time (s)',
    label='Dilated Time (Earth Observer)',
    color='blue'
)
plt.axhline(1.0, color='gray', linestyle='--', label='Proper Time (1s)')
plt.title("Time Dilation vs Velocity (Special Relativity)", fontsize=14)
plt.xlabel("Velocity (fraction of speed of light, c)")
plt.ylabel("Dilated Time (seconds)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## üåå Gravitational Time Dilation

This plot simulates how time is dilated **near a black hole**, where the intense gravity curves spacetime. The closer the observer to the event horizon (1 R‚Çõ), the slower time moves relative to a distant observer.

````python
plt.figure()
sns.lineplot(
    data=df_gravity,
    x='Distance in Rs',
    y='Dilated Time (s)',
    label='Dilated Time (Distant Observer)',
    color='darkorange'
)
plt.axhline(1.0, color='gray', linestyle='--', label='Proper Time (1s)')
plt.title("Gravitational Time Dilation vs Distance from Black Hole", fontsize=14)
plt.xlabel("Distance from Center (in Schwarzschild radii, R‚Çõ)")
plt.ylabel("Dilated Time (seconds)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````
